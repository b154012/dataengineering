```python
#2. Using b mode, write a cp tool. The requirements are as follows: 
#a. It can copy text, video, picture and other files.
#b. If the user's parameters are wrong, the tool should show how to use the print command correctly, such as usage: cp source_ file target_ file.
# Tip: you can use import sys, and then use sys. Argv to get the parameters following the script.

import os
import shutil
import sys

def cptool():
    if len(sys.argv) != 4:
        print("usage - cp source_file target_file")
    else:
        try:

            if sys.argv[1] != 'cp':
                raise TypeError("usage - cp source_file target_file")
            else:
                #shutil.copyfile(sys.argv[2], sys.argv[3])
                with open(sys.argv[2], 'rb') as rf, open(sys.argv[3], 'wb') as wf:
                    for line in rf:
                        wf.write(line)
        except IOError as e:
            print(e)
            print(f"File {sys.argv[2]} does not exist")
        except TypeError as e:
            print(e)
        else:
            print("Successfully copied the file")
        
cptool()
```

```python
#3. Write a program to read the CSV file and output the content of the file. 
import csv
import os

def readCSV():
    try:
        data = input("Enter your CSV file [name.csv]: ")
        data += '.csv'

        with open(data, 'rt') as file:
            data = dict(csv.reader(file))
            for key, value in data.items():
                if key == 'Name' or value == 'Code':
                    continue
                print(f"Country: {key} --- Code: {value}")
    except IOError as e:
        print(e)
    finally:
        print("Successfully read")
        
readCSV()
```

```python
import re
import csv
import os

def containsSpecialChar(password):
    special_char = False
    isContain = re.compile('[^0-9a-zA-Z]+')
    if isContain.search(password):
        print("The password cannot contain special characters!")
        special_char = True   
    return special_char

def saveUserRegistration(user_dictionary): 
    filename = "user.csv"
    
    if os.path.getsize(filename) == 0:
        try:
            with open(filename, 'a+', newline='') as csvfile:
                #used to have an extra blank row, removed it by using seek(0)- moves to the top line
                csvfile.seek(0)
                data = csv.writer(csvfile)
                for key, value in user_dictionary.items():
                    data.writerow([key, value])
        except IOError:
            print("IO error")
        finally:
            csvfile.close()
            print("Successfully saved the data into a file")
    else:
        try:
            with open(filename, 'a+', newline='') as csvfile:
                data = csv.writer(csvfile)
                for key, value in user_dictionary.items():
                    data.writerow([key, value])
        except IOError:
            print("IO error")
        finally:
            csvfile.close()
            print("Successfully saved the data into a file.")
        
def usernameExists(username):
    filename = "user.csv"
    index = 0
    
    if os.path.getsize(filename) == 0:
        print("No content yet.")
    else:
        try:
            with open(filename, 'r') as csvfile:
                data = dict(csv.reader(csvfile))
                if username in data:
                    print("Username Exist!")
                    return True
                else:
                    return False
        except IOError:
                print("IOError")
        finally:
                csvfile.close()
        
def cannotStartWithNumber(username):
    if re.match('[0-9]', username):
        print("Cannot start with a number. Try again!")
        return True
    else:
        return False
    
def lengthOverSix(username):
    if len(username) >= 6:
        return True
    else:
        print("The length of valid passwords should have at least 6"
              " alphanumeric characters and should not contain special characters.")
        return False
            
def main():
    num = -1

    while(num != 0):
        
        print("\n***************************** User Registration ****************************************")
        print("                             1. User Registration.                                        ")
        print("                             0. Exit                                                      ")
        print("******************************************************************************************")
        
        num = int(input("Enter your number: "))
        
        try:
            if num == 1:

                isUsername = False
                isPassword = False
                user_dictionary = {}
                
                while(not isUsername):
                    username = input("Enter your username: ")
                    if cannotStartWithNumber(username) or usernameExists(username):
                        continue
                    else:
                        isUsername = True
                        break

                while(not isPassword):
                    password = input("Enter your password: ")
                    # if contains special_char, it returns True but negates it
                    # if doesn't contain, it returns false and negates it. 
                    ''' Expects the not containsSpecialChar func = True '''
                    if lengthOverSix(password) and not containsSpecialChar(password):
                        isPassword = True
                        break
                    else:
                        continue
                '''If Satisfies the criteria, the save it into a file'''
                if (isUsername == isPassword):
                    user_dictionary = {}    
                    user_dictionary[username] = password.rstrip()
                    saveUserRegistration(user_dictionary)

            elif num == 0:
                print("Goodbye")
                break
            else:
                print("Please enter 0 or 1!.")
                main()
        except ValueError as e:
            print("Please enter positive integers only!.")
            main()
main()
```

```python
#5. urlist is a list of urls, such as: [‘http://abc.com‘, ‘http:///abc.com‘, ‘http://abcd.cm’ etc.]
# Define a function func (urlist) which does the below: 
#a. Open the URLs in turn.
#b. Print the content corresponding to the URL.
#c. If the URL cannot be opened, record the URL to the log file and skip to continue to visit the next URL.

import urllib.request
import urllib.parse
from urllib.error import URLError, HTTPError
import re
import logging

def func(urlist):
    
    for i in range(len(urlist)):

        req = urllib.request.Request(urlist[i])
        logging.basicConfig(filename='error_log.txt', level=logging.ERROR)
        try:
            with urllib.request.urlopen(req) as the_response:
                the_page = the_response.read()

        except URLError as e:
            if hasattr(e, 'reason'):
                print('Failed to reach a server.')
                logging.exception(e)
                continue
            elif hasattr(e, 'code'):
                print('The serve couldn\'t fulfill the request.')
                logging.exception(e)
                continue
        except Exception as e:
            print('Cannot find the URL!')
            logging.exception(e)
            continue
        finally:
            remove_html = re.compile(r'\W+') 
            contents = re.sub(remove_html, ' ', str(the_page))

            print(contents)
            

urlist = ['https://en.wikipedia.org/wiki/Cheese', 'https://evydacademy.org', 'https://www.reuters.com/world/asia-pacific/new-zealand-sees-success-curbing-delta-outbreak-new-cases-plunge-2021-09-03/']
func(urlist)
```
