```python
import os
import sys
import csv
import re
import getpass
from pathlib import Path
from datetime import datetime

class Student:
    
    '''Turns student dictionary into a class'''
    
    def __init__(self, student_dict, studentid, name, major):
        
        """Constructor"""
        self.___dict___ = student_dict
        self._studentid = studentid
        self._name = name
        self._major = major
 
    @property
    def studentid(self):
        return self._studentid

    @studentid.setter
    def studentid(self, studentid):
        self._studentid = studentid

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        self._name = name

    @property
    def major(self):
        return self._major

    @major.setter
    def major(self, major):
        self._major = major       

    ''''Return Dictionary'''   
    def __repr__(self):
        return "%s" % self.__dict__

    '''Return a list of strings'''
    def __str__(self):
        return f"\n------------------------Student Info------------------------\nStudent ID: {self.studentid}\nName: {self.name}\nMajor: {self.major}\n"      

def addUser(student_dict):
    '''Add user. Enter studentid and it does not exit
       Enter name and major
       Store the data into a dictionary and then turn a dictionary into a class
       return studentid'''
    
    flag = False  
    while(not flag):
        studentid = int(input("Enter your student ID: "))
        if (idExists(studentid) or digitOnly(studentid)):
            print("here")
            continue
        else:
            flag = True
            break

    studentid = int(studentid) 
    name = input("Enter your name: ")
    major = input("Enter your major: ")
    
    print("Successfully")
    student_dict[studentid] = {name, major}
    student = Student(student_dict, studentid, name, major)
    
    '''Save user data in student.csv'''
    saveUser(student_dict)
    '''display student using Student class'''
    displayUser(student)
        
    return studentid

def updateUser(student_dict, studentid):
    '''Updating user only changes name and major
    student automatically update their info using their own studentid
    also check if student has inputted their data in the database as in student.csv'''
    
    if len(student_dict) == 0:
        print("The dictionary is empty.")
    else:

        if studentid in student_dict:
            name = input("Update your name: ")
            major = input("Update your major: ")

            print("\nSuccessfully updated!")

            student_dict[studentid] = (name, major)
            student = Student(student_dict, studentid, name, major) 
            displayUser(student)

def searchUser(student_dict):
    '''Seach student info by iterating a dictionary'''
    
    if len(student_dict) == 0:
        print("No user is found.")
    else:
        studentid = int(input("Enter your student ID: "))
        for key, value in student_dict.items():
            if key == studentid:
                print("Found: ", student_dict)
                break
            else:
                print("Not found.")
        
def displayUser(student):
    '''Invoke student object to display the student info'''
    print(student)
        
def deleteUser(student_dict, studentid):
    '''Deleting user only occurs in the existing system.
    finding the key to be deleted and comparing with its key in dictionary'''
    
    if len(student_dict) == 0:
        print("No user is found.")
    else:
        
        for key, value in student_dict.items():
            if key == studentid:
                del student_dict[key]
                print("Successfully deleted a student with ID: ", key)
                break

    answer = input("\nWould you like to add student info again? [y/n]: ")

    '''Once successfully deleted, student asks again whether to input.'''
    if answer.lower() == 'y':
        student_menu()
    else:
        main()

def saveUser(student_dict): 
    '''Every user info is saved in student.csv. This file is a benchmark whether their info is saved.'''
        filename = "student.csv"           
        try:
            with open(filename, 'a+', newline='') as csvfile:
                data = csv.writer(csvfile)
                for key, value in student_dict.items():
                    print(list(value))
                    data.writerow([key, value])
        except IOError:
            print("IO error")
        finally:
            csvfile.close()
            print("Successfully saved the data into a file.")

def idExists(studentid):
    '''Using student.csv to detect whether a studentid exists in the file with the newly entered studentid'''
    filename = "student.csv"
    try:
        with open(filename, 'r') as csvfile:
            data = dict(csv.reader(csvfile))
            for k,v in data.items():
                if (int(k) == studentid):
                    print("ID Exists")
                    return True
            else:
                return False
    except IOError:
        print("IOError")
    finally:
        csvfile.close()

    #if this is not using csv to iterate over elements
    '''
    if studentid in student_dict:
        print("Student ID Exits!")
        return True
    else:
        return False
    '''

def saveUserRegistration(user_dictionary): 
    '''When user logins, their username and password are saved in user.csv'''
    filename = "user.csv"
    
    if os.path.getsize(filename) == 0:
        try:
            with open(filename, 'a+', newline='') as csvfile:
                #used to have an extra blank row, removed it by using seek(0)- moves to the top line
                csvfile.seek(0)
                data = csv.writer(csvfile)
                for key, value in user_dictionary.items():
                    data.writerow([key, value])
        except IOError:
            print("IO error")
        finally:
            csvfile.close()
            print("Successfully saved the data into a file")
    else:
        try:
            with open(filename, 'a+', newline='') as csvfile:
                data = csv.writer(csvfile)
                for key, value in user_dictionary.items():
                    data.writerow([key, value])
        except IOError:
            print("IO error")
        finally:
            csvfile.close()
            print("Successfully saved the data into a file.")

def saveIndividualUser(student_dict):  
    '''Every user has a choice to save their student info by entering a name file'''
    
    filename = input("Enter your filename (name.txt): ")
    myfile = Path(filename)
    myfile.touch(exist_ok=True)

    try:
        with open(myfile, 'w', newline='') as csvfile:
            data = csv.writer(csvfile)
            for key, value in student_dict.items():
                data.writerow([key, value])
    except IOError:
        print("IO error")
    finally:
        csvfile.close()
        print("Successfully saved the data into a file.")          

def usernameExists(username):
    '''This is to check the user authentication'''
    
    filename = "user.csv"
    
    try:
        with open(filename, 'r') as csvfile:
            data = dict(csv.reader(csvfile))
            if username in data:
                print("Username Exist!")
                return True
            else:
                return False
    except IOError:
        print("IOError")
    finally:
        csvfile.close()

def checkUsernamePassword(username, password):
    '''This is to chech the user authentication'''
    
    filename = "user.csv"
    
    try:
        with open(filename, 'r') as csvfile:
            data = dict(csv.reader(csvfile))
            for k, v in data.items():
                if (k == username) and (v == password):
                    print("Login!")
                    return True
            else:
                return False
    except IOError:
        print("IOError")
    finally:
        csvfile.close()
    
        
def cannotStartWithNumber(username):
    '''Enusres that username cannot start with a number'''
    
    if re.match('[0-9]', username):
        print("Cannot start with a number. Try again!")
        return True
    else:
        return False
    
def lengthOverSix(username):
    '''The length of username should be at least 6'''
    
    if len(username) >= 6:
        return True
    else:
        print("The length of valid passwords should have at least 6"
              " alphanumeric characters and should not contain special characters.")
        return False

def containsSpecialChar(password):
    '''Ensures that password does not contain special characters'''
    
    
    special_char = False
    isContain = re.compile('[^0-9a-zA-Z]+')
    if isContain.search(password):
        print("The password cannot contain special characters!")
        special_char = True   
    return special_char

def digitOnly(studentid):
    '''studentid cannot be negative integers'''

    if studentid < 0:
        print("Print a positive integer value only!")
        return True
    else:
        return False
    
def student_menu():
    
    num = -1
    time_now = datetime.now().time()
    student_dict = {}

    studentid = addUser(student_dict)
      
    while(num != 0):
        
        print("\n------------------------------ Student Management System --------------------------------")
        print("                                   1. Search                                                 ")
        print("                                   2. Delete                                                  ")
        print("                                   3. Update                                                  ")
        print("                                   4. Save                                                    ")
        print("                                   0. Exit                                                    ")
        print("----------------------------------------------------------------------------------------")
        print("\nCurrent time: ", time_now.strftime("%H:%M"), "\n")
        num = int(input("Enter your number: "))
        
        try:
            if num == 1:
                searchUser(student_dict)
            elif num == 2:
                deleteUser(student_dict, studentid)
            elif num == 3:
                updateUser(student_dict, studentid)
            elif num == 4:
                saveIndividualUser(student_dict)
            elif num == 0:
                sys.exit(1)
            else: 
                print("Please enter between 1 and 3.")
                main()
        except ValueError as e:
            print("Please enter positive integers only!")
            break

def main():
    num = -1
    time_now = datetime.now().time()

    while(num != 0):
        
        print("\n***************************** Student Management ****************************************")
        print("                             1. Login                                                     ")
        print("                             2. Register                                                  ")
        print("                             0. Exit                                                       ")
        print("*****************************************************************************************")
        print("\nCurrent time: ", time_now.strftime("%H:%M"), "\n")
        
        num = int(input("Enter your number: "))
        
        try:
            if num == 1:
                username = input("Enter your username: ")
                password = getpass.getpass(prompt = "Enter your password: ")
                if checkUsernamePassword(username, password):
                    print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~Welcome to Student Management~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
                    student_menu()
                    
                else:
                    print("Username or/and password is incorrect!")
            elif num == 2:
                
                isUsername = False
                isPassword = False
                user_dictionary = {}
                
                while(not isUsername):
                    username = input("Enter your username: ")
                    if cannotStartWithNumber(username) or usernameExists(username):
                        continue
                    else:
                        isUsername = True
                        break

                while(not isPassword):
                    password = input("Enter your password: ")
                    # if contains special_char, it returns True but negates it
                    # if doesn't contain, it returns false and negates it. 
                    ''' Expects the not containsSpecialChar func = True '''
                    if lengthOverSix(password) and not containsSpecialChar(password):
                        isPassword = True
                        break
                    else:
                        continue
                '''If Satisfies the criteria, the save it into a file'''
                if (isUsername == isPassword):
                    user_dictionary = {}    
                    user_dictionary[username] = password.rstrip()
                    saveUserRegistration(user_dictionary)

                #go to menu()
                student_menu()

            elif num == 0:
                print("Goodbye")
                sys.exit(1)
            else:
                print("Please enter 0 or 1!.")
                main()
        except ValueError as e:
            print("Please enter positive integers only!.")
            main()

if __name__ == '__main__':
    main()
```
