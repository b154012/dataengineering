### 2.Write a program to place furniture in a house:

Explanation:

There are two functions such as remaining area and total area. Remaing area returns the total area of the house minus the toal area of the furniture. The total area calculates the sum of the furniture area covers.

Furniture class inherits from its base class such as house to get member instances.
Furniture class overrides total_area

```python
class House:
    def __init__(self, house_name, total_area, furniture_list=None, area_covers=None):
        self._house_name = house_name
        self._total_area = total_area
        self._furniture_list = list(furniture_list)
        self._area_covers = list(area_covers)

    @property
    def house_name(self):
        return self._house_name

    @house_name.setter
    def house_name(self, house_name):
        self._house_name = house_name

    @property
    def total_area(self):
        return self._total_area

    @total_area.setter
    def total_area(self, total_area):
        self._total_area = total_area

    @property
    def furniture_list(self):
        return self._furniture_list

    @furniture_list.setter
    def furniture_list(self, furniture_list):
        self._furniture_list = furniture_list

    @property
    def area_covers(self):
        return self._area_covers

    @area_covers.setter
    def area_covers(self, area_covers):
        self._area_covers = area_covers

    #remaining area
    def remaining_area(self):
        
        if not self._area_covers:
            return self._total_area
        else:
            return (self._total_area) - (Furniture.total_area(self))

    def __str__(self):
        return f"The name of the house is {self._house_name} with a size of {self._total_area} m2.\nIt has a list of furnitures such as {self._furniture_list}."

class Furniture(House):

    def __init__(self, house_name, total_area, furniture_list=None, area_covers=None):
        super().__init__(house_name, total_area, furniture_list, area_covers)

    def total_area(self):
        sum = 0
        for i in self._area_covers:
            sum += i 
            
        if sum > self._total_area:
            return 0
        else:
            return sum
    
def main():
    
    '''a. House has a list of house type, total area and furniture name. 
       b. New house has no furniture.
       c. Furniture info are given below:'''
    house_name = "Two Storey"
    total_area = 400
    furniture = ["Bed", "Cupboard", "Table"]
    area_covers = [10, 2, 5]
    
    '''Print the house name, total area, remaining area, furniture'''
    house = Furniture(house_name, total_area, furniture,  area_covers)
    print(house, "The total area is", house.total_area(), "m2. The remaining area is", house.remaining_area(), "m2.")
    
if __name__ == '__main__':
    main()
```
