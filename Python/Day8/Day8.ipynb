```python
#5. Abstract apple and banana into a class. Provide basic properties and basic methods.
# Then, create several different objects through classes. Print their properties and call their methods. 

from abc import ABC, abstractmethod

class Fruit(ABC):

    @abstractmethod
    def color(self):
        pass
    
    def properties(self):
        pass
    
class Apple(Fruit):
    
    def color(self):
        return f"The color of apple is red."
    
    def properties(self):
        return f"The properities of apple include:\n1.Lower High Cholesterol\n2. Enhance Blood Pressure\n3. Rich in Antioxidants"

class Banana(Fruit):
    
    def color(self):
        return f"The color of banana is yellow."
        
    def properties(self):
        return "The properties of banana include:\n1. Maintain blood sugar levels.\n2. Improve digestive systems.\n3. Aid weight loss."
        
def main():
    apple = Apple()
    print(apple.color())
    print(apple.properties())
    
    print()
    
    banana = Banana()
    print(banana.color())
    print(banana.properties())
    
if __name__ == '__main__':
    main()
```

```python
#11. Write a FileObject class which wraps a file object 
# and confirms that the file is automatically closed when the object is deleted.

class FileObject():
    
    def __init__(self, filename):
        try:
            print("__init__ is running...")
            self.filename = filename
            self.fileobj = open(filename, 'w')
        except IOError as e:
            print(e)
        
    def writeData(self, data):
        
        try:
            self.fileobj.write(data)
            print("Data is created.")
        except IOError as e:
            print(e)
            
    def __del__(self):
        if self.fileobj != 0:
                print("File is automatically closed.")
                print("Object is deleted using __del__")
        
def main():
    fileobject = FileObject("data.txt")
    fileobject.writeData("I am Zul learning")

if __name__ == '__main__':
    main()
```

```python
#12. Define a base class called Shape. 
#a. Private data consist of coordinate points x, y, from which the Rectangle class and Circle class are derived. 
#They all have the member function getarea() to calculate the area. 
#b. The data for the derived class Rectangle class are the length and width of rectangle
#c. The date for the derived class Circle class is the radius of the circle. 
#d. The Rectangle class inherits all the properties of the class Square. Define and test this Rectangle class.
#Input three groups of data, namely the length and width of the rectangle, the radius of the circle 
#and the side length of the square. 
#e. Output three groups of data, namely the areas of the rectangle, the circle and the square. (π = 3.14)
#f. Output description: Three output lines for the areas of the rectangle, the circle and the square respectively.

import math
from abc import ABC, abstractmethod

class Shape(ABC):
    
    def __init__(self, x, y):
        self._x = x
        self._y = y
    
    @property
    def x(self):
        return self._x
    
    @property
    def y(self):
        return self._y
    
    @x.setter
    def x(self, x):
        self._x = x

    @y.setter
    def y(self, y):
        self._y = y
     
    @abstractmethod
    def getArea(self):
        pass
    
    def __str__(self):
        return f"x: {self._x}\ny: {self._y}"
    
class Square(Shape):
    
    def __init__(self, x, y):
        super().__init__(x, y)
        
    def getArea(self):
        result = self._x * self._y
        return "{:.2f}".format(result)
        
class Rectangle(Square):
    
    def __init__(self, x ,y):
        super().__init__(x, y)
        
    def getArea(self):
        result = 1/2 * self._x * self._y
        return "{:.2f}".format(result)
    
class Circle(Shape):
    
    def __init__(self, x, y):
        super().__init__(x, y)
        
    def getArea(self):
        result = math.pi * self._x * self._y
        return "{:.2f}".format(result)

def main():
    
    length = int(input("Enter the length of the rectangle: "))
    width = int(input("Enter the width of the rectangle: "))
    rectangle = Rectangle(length,width)
    
    radius = int(input("Enter the radius of the circle: "))
    circle = Circle(radius, radius)
    
    size_length = int(input("Enter the side length of the Square: "))
    square = Square(size_length, size_length)
    
    print("\nThe area of rectangle is ", rectangle.getArea())
    print("The area of circle is ", circle.getArea())
    print("The area of square is ", square.getArea())
    
if __name__ == '__main__':
    main()
```

```python
# 14. Define a class to convert Centigrade to Fahrenheit (Fahrenheit = Centigrade * 1.8 + 32). 
class Convert:
    
    def __init__(self, centigrade):
        self._centigrade = centigrade
        
    @property
    def centigrade(self):
        return self._centigrade
    
    @centigrade.setter
    def centigrade(self, centigrade):
        self._centigrade = centigrade
        
    def convertFahrenheit(self):
        result = self._centigrade * 1.8 + 32
        return "{:.2f}".format(result)

def main():
    try:
        num = int(input("Enter your centigrade: "))
        convert = Convert(num)
        print("The result of converting", num,"Centigrade to Fahrenheit is ", convert.convertFahrenheit())
    except ValueError as e:
        print(e)
    
if __name__ == '__main__':
    main()
```
